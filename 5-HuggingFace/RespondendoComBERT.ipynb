```python

import transformers
import torch
from transformers import BertTokenizer, BertForQuestionAnswering
```

```python
model_name = "bert-large-uncased-whole-word-masking-finetuned-squad"
```

```python
model = BertForQuestionAnswering.from_pretrained(model_name)
```

```python
tokenizer = BertTokenizer.from_pretrained(model_name)
```

# Ebeddings

```python
question = "Quando que o primeiro dvd foi lançado? e qual foi?" 
answer_document = "O DVD foi lançado oficialmente em 1º de novembro de 1996, no Japão, com o leitor Toshiba SD-3000. Os primeiros filmes lançados incluíram títulos como Twister, Blade Runner e Eraser. A tecnologia chegou aos EUA em 1997 e ao Brasil por volta de 1997/1998, com o filme Era uma Vez na América. "
```

```python

encoding = tokenizer(
    question,
    answer_document,
    return_tensors="pt"
)
```

```python
inputs = encoding ['input_ids']
sentence_embedding = encoding['token_type_ids']
tokens = tokenizer.convert_ids_to_tokens(inputs)
```
